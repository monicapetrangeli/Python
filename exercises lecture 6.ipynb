## Pandas I

And how do we know about missing elements and handle them?

```python
# What about missing values? Missing values are usually represented as NaN (Not a Number)
import numpy as np
data = {'Name': ['Amelie', 'Edgar', 'Carlos', 'Victor'],
        'Age': [24, 27, 22, 32],
        'Country': ['FR', np.nan, 'ES', 'GE']}
df = pd.DataFrame(data)
df
```

```python
# we can simply know which ones present missing values by using isna or isnull
df.isna() # at the whole dataframe level
```

```python
# and at the column level
df['Country'].isnull() # same as .isna()
```

```python
# Simple way of filling these ones is achieved by calling fillna
df.fillna('RE')
```

```python
# There's also possible thing you can do, use the mode (most common value)
#df['Country'].mode()
df.fillna(df['Country'].mode()[0])
```

```python
# IMPORTANT: if we check again the dataframe, we will see that the dataframe keeps having NaNs
# This is because the intented behaviour is usually to keep the DataFrame immutable
# If you want to actually modify you can pass ot fillna, inplace=True
df
```

```python
# A more drastic operation is to drop any column or row with missing values using
# here, we will modify the original dataframe and drop the rows and cols with nans
df.at[0, 'Age'] = np.nan
df
```

```python
df.dropna()
```

## Pandas I

One of the handy tools you have in **Pandas** is **groupby**. It allows you to group by any unique attributes within a column and get statistics:
- count
- mean
- sum
- min and max
- multiple aggregations
- group using multiple columns

```python
data = {
    'Name': ['Amelie', 'Edgar', 'Carlos', 'Victor', 'Sofia', 'Jin', 
             'Marta', 'Ali', 'Emily', 'Ravi', 'Chen', 'Fatima', 'Saham'],
    'Age': [24, 27, 22, 32, 29, 31, 28, 26, 23, 34, 25, 30, 26],
    'Country': ['FR', 'US', 'ES', 'GE', 'PT', 'KR',
                'ES', 'AE', 'US', 'IN', 'CN', 'AE', 'AE'],
    'Salary': [70000, 110000, 65000, 82000, 48000, 39000, 45000,
            90000, 97000, 31000, 49000, 85000, 80000]   
}

df = pd.DataFrame(data)
df
```

```python
# let's count how many we have per country
df.groupby('Country').count()
```

```python
df.groupby('Country')['Age'].mean() # now let's know the average age per country
```

```python
df.groupby('Country')['Salary'].sum() # total sum of their salaries
```

```python
#Â Multiple Aggregations
df.groupby('Country').agg({'Salary': ['min', 'max'], 'Age': 'mean'})
```

```python
# grouping using more than one column
df.groupby(['Country', 'Age'])['Salary'].count()
```

Home exercises for Netflix:

1. Is there any missing rating?
2. How many films in 2021 correspond to your country?
3. What's the number of movies in 2020 with full information?
4. Give me the year with more titles,
5. and what has been the average in terms of releases from 2010. 

And for Titanic:

1. Calculate Gender-Based Survival Percentage

2. Calculate Survival Percentage Grouped by Gender and Class

```python
import pandas as pd

#importing the dataset
netflix=pd.read_csv('netflix_titles.csv')

#missing data in ratings
print(f' there are {netflix['rating'].isna().sum()} missing values in the rating column')
#there are 4 missing data in rating

#my home country is Italy. Checking how many movies in 2021 have country==Italy
Italy_2021=netflix[(netflix['release_year']==2021) & (netflix['country']=='Italy')]
print(f'There are {len(Italy_2021)} movies in 2021 corresponding to Italy')

#counting the number of movies in 2020 with full information
#first cleaning all missing values
no_missing=netflix.dropna()
movies_2020=no_missing[no_missing['release_year']==2020]
print(movies_2020)

#getting which year had the most films released
release=dict()
for row in range(len(netflix)):
    year=netflix.iloc[row]['release_year']
    if year not in release:
        release[year]=1
    else:
        release[year]+=1

most_movies_year=max(release) 
print(f'{most_movies_year} was the year with most movies released')

#finsing the average released from 2010 onward
#finsing first how many movies where released per year from 2010 onward using groupby and count
releases_per_year=netflix[netflix['release_year'] >= 2010].groupby('release_year').count()

#show_id will be used to find the overall mean since it has no missing vlaues within the dataframe
print(f'the average releases per year since 2010 onward is {releases_per_year['show_id'].mean()} movies per year, rounded to {round(releases_per_year['show_id'].mean())}')
```

```python
#importing the dataset
titanic=pd.read_csv('train_and_test2.csv')

#grouping the data based on gender and survival
grouped_gender=titanic.groupby(['Sex','2urvived']).count()
#dividing the number of male/female passengers survived (==1) divided by the total number of male/female passengers counted through passengerid because of the lack of missing data in it
survival_female=(grouped_gender.iloc[1,1]/titanic[titanic['Sex']==0]['Passengerid'].count())*100
survival_male=(grouped_gender.iloc[3,1]/titanic[titanic['Sex']==1]['Passengerid'].count())*100
print(f'{survival_female} of female passengers survived while {survival_male} of male passengers survived.')

#grouping the data based on gender and class
grouped_class=titanic.groupby(['Sex','2urvived','Pclass']).count()
#dividing the number of male/female passengers in each class which survived (==1) divided by the total number of male/female passengers in that class counted through passengerid because of the lack of missing data in it
survival_female_1=(grouped_class.iloc[3,1]/titanic[(titanic['Sex']==0)&(titanic['Pclass']==1)]['Passengerid'].count())*100
survival_female_2=(grouped_class.iloc[4,1]/titanic[(titanic['Sex']==0)&(titanic['Pclass']==2)]['Passengerid'].count())*100
survival_female_3=(grouped_class.iloc[5,1]/titanic[(titanic['Sex']==0)&(titanic['Pclass']==3)]['Passengerid'].count())*100
survival_male_1=(grouped_class.iloc[9,1]/titanic[(titanic['Sex']==1)&(titanic['Pclass']==1)]['Passengerid'].count())*100
survival_male_2=(grouped_class.iloc[10,1]/titanic[(titanic['Sex']==1)&(titanic['Pclass']==2)]['Passengerid'].count())*100
survival_male_3=(grouped_class.iloc[11,1]/titanic[(titanic['Sex']==1)&(titanic['Pclass']==3)]['Passengerid'].count())*100

print(f'{survival_female_1} females travelled in class 1 survived, {survival_female_2} in class 2, and {survival_female_3} in class 3')
print(f'{survival_male_1} males travelled in class 1 survived, {survival_male_2} in class 2, and {survival_male_3} in class 3')
```
